```python
# Agent-Based Workflow Automation System

class WorkflowAgent:
    def __init__(self):
        self.tasks = []

    def receive_task(self, task_name):
        print(f"\nTask Received: {task_name}")
        self.tasks.append(task_name)

    def process_task(self, task_name):
        print(f"Processing Task: {task_name}")

        if task_name.lower() == "generate report":
            self.generate_report()

        elif task_name.lower() == "send email":
            self.send_email()

        elif task_name.lower() == "backup data":
            self.backup_data()

        else:
            print("No predefined workflow found.")

    def generate_report(self):
        print("Generating report...")
        print("Report generated successfully.")

    def send_email(self):
        print("Sending email notification...")
        print("Email sent successfully.")

    def backup_data(self):
        print("Backing up data...")
        print("Data backup completed.")

    def execute_workflow(self):
        print("\n----- Workflow Execution Started -----")

        for task in self.tasks:
            self.process_task(task)

        print("\n----- Workflow Execution Completed -----")


# Main Program
if __name__ == "__main__":
    agent = WorkflowAgent()

    # Sample Tasks
    agent.receive_task("Generate Report")
    agent.receive_task("Send Email")
    agent.receive_task("Backup Data")

    agent.execute_workflow()
```
